# Let's visualize our data!


# Scatter plot

```python
plt.scatter('adults','stays_in_weekend_nights',data=df)
```

```python
# canceled=1.Not canceled=0
sns.countplot(data=df, x='is_canceled')
plt.show()
```

```python
sns.countplot(x='hotel',hue='is_canceled',data=df)
plt.title("number of cancellations")
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,cmap="summer")
```

# Box plot

```python
#Boxplot is another simple graphical method that can be used to summarize 
the distribution of the data.
ax = sns.boxplot(x=df['is_canceled'],y=df['lead_time'],hue=df['hotel'])
ax.set_title('Lead_Time_Box_Plot')
plt.xlabel('lead time')
plt.ylabel('is canceled')
plt.show()
```

# Histogram

Let's use histogram to visualize the distributions of the arrival date week hotel reservations number count.
The histogram is a way to show the data distribution in the frequency 
table.

```python
ax = plt.hist(x=df['arrival_date_week_number'],bins=np.arange(55)-0.5,facecolor='blue',alpha=0.5)
plt.xlabel('arrival_date_week_number')
plt.ylabel('count')
plt.title('arrival_date_week_number_count')
plt.xticks(range(1,54,10))

plt.show()
```
